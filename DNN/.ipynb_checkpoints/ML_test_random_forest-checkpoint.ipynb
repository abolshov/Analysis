```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
data_signal = pd.read_csv("data_signal.csv")
data_backgr = pd.read_csv("data_backgr.csv")
```

```python
data_signal.head()
```

```python
data_backgr.head()
```

```python
list(data_backgr.columns) == list(data_signal.columns)
```

```python
data_signal[data_signal.dnn_truth_value != 0]
```

```python
data_backgr[data_backgr.dnn_truth_value != 0]
```

```python
data_backgr.dnn_truth_value.unique()
```

```python
data_signal.dnn_truth_value.unique()
```

```python
frames = [data_signal, data_backgr]
data = pd.concat(frames)
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
from sklearn.utils import shuffle
data = shuffle(data)
data.head(30)
```

```python
y = data.dnn_truth_value
X = data.drop('dnn_truth_value', axis=1)
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
clf_rf = RandomForestClassifier(random_state=0,
                               n_estimators=20, 
                               max_depth=20,
                               min_samples_split=50,
                               min_samples_leaf=50)
```

```python
clf_rf.fit(X_train, y_train)
```

```python
clf_rf.score(X_test, y_test)
```

```python
imp = pd.DataFrame(clf_rf.feature_importances_, index=X_train.columns, columns=['importance'])
imp_hist = imp.sort_values('importance').plot(kind='barh', figsize=(10, 15))
fig = imp_hist.get_figure()
fig.savefig("imp_hist.png")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(60,60))

corr_signal = data_signal.drop(['met_E', 'met_pz', 'dnn_truth_value'], axis=1).corr()
signal_hmap = sns.heatmap(corr_signal, 
                          xticklabels=corr_signal.columns.values,
                          yticklabels=corr_signal.columns.values,
                          annot=True, 
                          cmap='viridis',
                          ax=ax)
plt.yticks(rotation=0)
fig = signal_hmap.get_figure()
fig.savefig("signal_hmap.png")
```

```python
fig, ax = plt.subplots(figsize=(60,60))
corr_backgr = data_backgr.drop(['met_E', 'met_pz', 'dnn_truth_value'], axis=1).corr()
x_labels = corr_backgr.columns.values
y_labels = corr_backgr.columns.values
backgr_hmap = sns.heatmap(corr_backgr, 
                          xticklabels=x_labels,
                          yticklabels=y_labels,
                          annot=True, 
                          cmap='viridis',
                          ax=ax)
# plt.yticks(rotation=0) 
backgr_hmap.set_yticklabels(backgr_hmap.get_yticklabels(), rotation = 0)
fig = backgr_hmap.get_figure()
fig.savefig("backgr_hmap.png")
```
